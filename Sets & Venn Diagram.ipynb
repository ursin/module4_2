```python
# https://github.com/konstantint/matplotlib-venn
import matplotlib.pyplot as plt
from matplotlib_venn import *

plt.figure(figsize=(10,10))

set1 = set(['A', 'B', 'C', 'D'])
set2 = set(['B', 'C', 'D', 'E'])
set3 = set(['C', 'D',' E', 'F', 'G'])

venn = venn3((set1, set2, set3), ('Set 1', 'Set 2', 'Set 3'))
c = venn3_circles(subsets=(set1, set2, set3), linestyle='dashed')

for text in venn.set_labels:
    text.set_fontsize(26)
for text in venn.subset_labels:
    text.set_fontsize(24)

venn.get_label_by_id('100').set_text(', '.join(set1 - set2 - set3))
venn.get_label_by_id('110').set_text(', '.join(set1 & set2 - set3))
venn.get_label_by_id('010').set_text(', '.join(set2 - set3 - set1))
venn.get_label_by_id('101').set_text(', '.join(set1 & set3 - set2))
venn.get_label_by_id('111').set_text(', '.join(set1 & set2 & set3))
venn.get_label_by_id('011').set_text(', '.join(set2 & set3 - set1))
venn.get_label_by_id('001').set_text(', '.join(set3 - set2 - set1))

plt.show()
```

```python
plt.figure(figsize=(10,10))

set1 = set(['A', 'B', 'C', 'D'])
set2 = set(['C', 'D', 'E', 'F'])

venn = venn2((set1, set2), ('Set 1', 'Set 2'))
c = venn2_circles(subsets=(set1, set2), linestyle='dashed')

for text in venn.set_labels:
    text.set_fontsize(26)
for text in venn.subset_labels:
    text.set_fontsize(24)

venn.get_label_by_id('10').set_text(', '.join(set1 - set2))
venn.get_label_by_id('11').set_text(', '.join(set1 & set2))
venn.get_label_by_id('01').set_text(', '.join(set2 - set1))

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib_venn import venn3, venn3_circles
plt.figure(figsize=(6,6))
v = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels = ('A', 'B', 'C'))
v.get_patch_by_id('100').set_alpha(1.0)
v.get_patch_by_id('100').set_color('white')
v.get_label_by_id('100').set_text('Unknown')
v.get_label_by_id('A').set_text('Set "A"')
c = venn3_circles(subsets=(1, 1, 1, 1, 1, 1, 1), linestyle='dashed')
c[0].set_lw(1.0)
c[0].set_ls('dotted')
plt.title("Sample Venn diagram")
plt.annotate('Unknown set', xy=v.get_label_by_id('100').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
             ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'))
plt.show()
```
